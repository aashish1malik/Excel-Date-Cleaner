```python
# %% [markdown]
# ## 1. Data Loading & Initial Inspection

# %%
# Load data
df = pd.read_excel(r'C:\Users\user\Desktop\genai project\project 1\student.xlsx')

# Initial inspection
print(f"Dataset shape: {df.shape}")
df.head()

# %%
# Basic info
df.info()
```

```python
# %% [markdown]
# ## 2. Data Cleaning

# %%
# Handle missing/inconsistent values
print("Unique values in 'Scholarship':", df['Scholarship'].unique())

# Clean 'Scholarship' column
df['Scholarship'] = df['Scholarship'].replace({'None': np.nan, '50%': 50, '75%': 75, '100%': 100, '25%': 25})
df['Scholarship'] = pd.to_numeric(df['Scholarship'])

# Fix inconsistent values in other columns
text_columns = ['Reading', 'Notes', 'Listening_in_Class']
for col in text_columns:
    df[col] = df[col].replace({6: 'Yes', '6': 'Yes'})  # Assuming 6 was data entry error

# %%
# Check cleaned data
df.describe(include='all')
```

```python
# %%
# Grade distribution
plt.figure(figsize=(10,5))
sns.countplot(data=df, x='Grade', order=sorted(df['Grade'].unique()))
plt.title('Grade Distribution')
plt.show()
```

```python
# %%
# Numerical features analysis
num_cols = ['Student_Age', 'Weekly_Study_Hours', 'Scholarship']
df[num_cols].hist(figsize=(12,4), bins=15)
plt.tight_layout()
plt.show()
```

```python
# Categorical features analysis
cat_cols = ['Sex', 'High_School_Type', 'Additional_Work', 'Sports_activity']
fig, axes = plt.subplots(2, 2, figsize=(12,8))
for col, ax in zip(cat_cols, axes.flatten()):
    sns.countplot(data=df, x=col, ax=ax)
plt.tight_layout()
plt.show()
```

```python
# ## 4. Feature Engineering

# %%
# Encode categorical variables
label_encoders = {}
categorical_cols = ['Sex', 'High_School_Type', 'Transportation', 'Attendance', 
                   'Reading', 'Notes', 'Listening_in_Class', 'Project_work']

for col in categorical_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
    label_encoders[col] = le

# Create binary flags
binary_cols = ['Additional_Work', 'Sports_activity']
df[binary_cols] = df[binary_cols].replace({'Yes': 1, 'No': 0})

# %%
# Check transformed data
df.head()
```

RAG = Retrieval-Augmented Generation

Itâ€™s a technique where LLMs (like GPT-4) are combined with external knowledge sources (like documents, databases) to answer questions more accurately and with updated information.

ðŸ¤” Why is it needed?
LLMs like GPT-4 are powerful, but:


* They canâ€™t access real-time or personal documents unless theyâ€™re trained on them.
* Their knowledge is limited to their training data (and cutoff date).
* They may hallucinate when they don't have enough context.

So, how do we solve this?
ðŸ‘‰ By â€œretrievingâ€ relevant info first, and then generating answers with it!

ðŸ”§ How RAG Works (Simplified Flow):

* User asks a question â†’ â€œWhat is Generative AI?â€
* System searches relevant documents (e.g., PDFs, websites, files) using vector similarity search
* It retrieves top matching chunks
* These chunks are given as context to the LLM
* The LLM generates an answer based on that context


ðŸ§  Think of RAG as:
â€œðŸ¤– A smart assistant that first reads your files before answering your questions.â€

ðŸ§° RAG = Retriever + Generator


* Component	-----------------------RoleRetriever
* Finds     -----------------------relevant information using vector search
* Generator -----------------------Uses LLM (like GPT) to answer based on that info



âœ… Benefits of RAG
Up-to-date answers (even from personal data)


* Better context = fewer hallucinations
* Doesnâ€™t require fine-tuning big models
* Can work with private or domain-specific knowledge

```python
from sentence_transformers import SentenceTransformer
from transformers import pipeline
import faiss
import numpy as np
```

```python
# ðŸ“ Documents
documents = [
    "Python is a versatile language.",
    "RAG is used to improve LLMs with external knowledge.",
    "FAISS helps in fast vector similarity search.",
    "Transformers are powerful models for NLP tasks.",
    "Embeddings are numeric representations of text."
]
```

```python
# ðŸ”  Step 1: Create Embeddings
model = SentenceTransformer('all-MiniLM-L6-v2')
doc_embeddings = model.encode(documents)
```

```python
# ðŸ” Step 2: FAISS Index
dimension = doc_embeddings.shape[1]
index = faiss.IndexFlatL2(dimension)
index.add(np.array(doc_embeddings))
```

```python
# â“ Step 3: User Query
query = "How does RAG help language models?"
query_embedding = model.encode([query])
D, I = index.search(np.array(query_embedding), k=2)
```

```python
# ðŸ“š Step 4: Retrieve Context
retrieved_docs = [documents[i] for i in I[0]]
context = "\n".join(retrieved_docs)
```

```python
# ðŸ¤– Step 5: Use Local Model for Answer Generation
qa_pipeline = pipeline("text-generation", model="distilgpt2")
```

```python
# Basic prompt construction
prompt = f"Context:\n{context}\n\nQuestion: {query}\nAnswer:"

# Generate answer
output = qa_pipeline(prompt, max_length=100, do_sample=True)[0]['generated_text']

# âœ… Output
print("ðŸ”Ž Context:")
print(context)
print("\nðŸ’¬ Generated Answer:\n", output)
```
